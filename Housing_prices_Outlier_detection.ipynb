```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
df_train = pd.read_csv("../data/train.csv")
df_test=pd.read_csv("../data/test.csv")

# Drop Id columns
df_train = df_train.drop(['Id'],axis=1)
df_test = df_test.drop(['Id'],axis=1)

print("Training Dataframe shape:\t ",df_train.shape,"\nTesting Dataframe shape:\t",df_test.shape)
```

```python
continuous_cols = ["LotFrontage","LotArea","MasVnrArea","BsmtFinSF1","BsmtFinSF2", "BsmtUnfSF",
                   "TotalBsmtSF","1stFlrSF","2ndFlrSF","GrLivArea","GarageArea","WoodDeckSF",
                   "OpenPorchSF","EnclosedPorch","3SsnPorch","ScreenPorch","PoolArea","MiscVal","LowQualFinSF"]
```

```python
df_train.shape
```

```python
#set theme for seaborn plot
sns.set()
fig, axes =plt.subplots(10,2,figsize=(20, 40))
for cols in range(19):
    column = continuous_cols[cols]
    sns.regplot(x=column, y="SalePrice", data=df_train,line_kws={"color": "orange"},ax=axes[int(cols/2)][cols%2]);
#     sns.scatterplot(x=column, y='SalePrice', data=df,ax=axes[int(cols/2)][cols%2],line_kws={"color": "orange"})
```
